```python
# Supondo que df seja o seu DataFrame
completed = df_final_web_data[df_final_web_data['process_step'] == 'step_3']
completed

# Agora pegamos os visitor_ids Ãºnicos que chegaram ao step_3
visitors_completed = completed['visitor_id'].unique()
visitors_completed

# Se quiser o DataFrame sÃ³ com os que completaram
df_completed = df_final_web_data[df_final_web_data['visitor_id'].isin(visitors_completed)]
df_completed
```

```python
# Por exemplo, checando se cada visitor_id passou por todos os steps
required_steps = {'step_1', 'step_2', 'step_3'}
visitor_steps = df_final_web_data.groupby('visitor_id')['process_step'].apply(set)

# Visitantes que passaram por todos os steps
completed_visitors = visitor_steps[visitor_steps.apply(lambda x: required_steps.issubset(x))].index

# Filtrando o DataFrame original
df_completed = df_final_web_data[df_final_web_data['visitor_id'].isin(completed_visitors)]
df_completed
```
